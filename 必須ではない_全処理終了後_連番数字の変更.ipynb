```python
import os
import re
import tkinter as tk
from tkinter import filedialog
from pathlib import Path
```

選択したフォルダ内にある、三桁の連番ファイルのスタート番号を変えて連番数字を振り直す。
ただし、ファイル名は108_190℃_320_1.0mmの構成になっている必要がある。

```python

# ==========================================
# 設定
# ==========================================
# 新しい連番の開始番号を指定してください
START_NUMBER = 1

# 連番の桁数を指定してください（例: 3桁の場合は 3）
ZERO_PADDING = 3
# ==========================================

def main():
    # Jupyter Notebook上でダイアログが裏に隠れないようにするための設定
    root = tk.Tk()
    root.withdraw() # メインウィンドウを非表示にする
    root.attributes('-topmost', True) # ダイアログを最前面に表示
    
    # フォルダ選択ダイアログを表示
    target_dir = filedialog.askdirectory(title="連番を振り直すフォルダを選択してください")
    root.destroy()
    
    if not target_dir:
        print("フォルダが選択されませんでした。処理を中止します。")
        return
        
    target_path = Path(target_dir)
    
    # 正規表現: 行頭の数字(1文字以上)と、アンダースコア以降の文字列をグループ化して抽出
    # 例: "108_190℃_320_1.0mm.txt" -> group(1)="108", group(2)="_190℃_320_1.0mm.txt"
    pattern = re.compile(r'^(\d+)(_.*)$')
    
    # 処理対象ファイルを収集
    target_files = []
    
    for file_path in target_path.iterdir():
        if file_path.is_file():
            match = pattern.match(file_path.name)
            if match:
                # 元の連番を数値として取得（ソート用）
                original_number = int(match.group(1))
                suffix = match.group(2)
                target_files.append({
                    'original_path': file_path,
                    'original_number': original_number,
                    'suffix': suffix
                })
                
    if not target_files:
        print("対象となるファイル（先頭が「数字＋アンダースコア」の形式）が見つかりませんでした。")
        return
        
    # 元の連番（数値）で昇順にソートし、順番を一切変えないようにする
    target_files.sort(key=lambda x: x['original_number'])
    
    # --- リネーム処理 ---
    # 既存のファイル名と新しいファイル名が被って上書きされるのを防ぐため、
    # 一旦すべて「一時的なファイル名」に変更してから、目的のファイル名に変更する2段階方式をとります。
    
    temp_files = []
    
    # 第1段階: 一時ファイル名にリネーム
    for i, file_data in enumerate(target_files):
        new_number = START_NUMBER + i
        # 指定した桁数でゼロ埋め（例: 1 -> "001"）
        new_number_str = str(new_number).zfill(ZERO_PADDING)
        new_filename = f"{new_number_str}{file_data['suffix']}"
        
        # 名前衝突回避用の一時プレフィックス
        temp_filename = f"__temp_rename_{i}_{new_filename}"
        temp_path = target_path / temp_filename
        
        file_data['original_path'].rename(temp_path)
        temp_files.append((temp_path, target_path / new_filename))
        
    # 第2段階: 最終的な名前にリネーム
    print(f"--- 選択フォルダ: {target_dir} ---")
    for temp_path, final_path in temp_files:
        temp_path.rename(final_path)
        print(f"変更完了: {final_path.name}")
        
    print(f"\n合計 {len(temp_files)} 個のファイル名の連番を振り直しました。")

# 実行
if __name__ == "__main__":
    main()
```
