```python
import tkinter as tk
from tkinter import filedialog
from pathlib import Path
```

ファイル名の末尾にそのファイルが入っているひとつ前のフォルダ名を付け足す。

```python

def select_directory():
    """ダイアログを開いてフォルダを選択する関数"""
    # Tkinterのメインウィンドウを作成し、非表示にする
    root = tk.Tk()
    root.withdraw()
    
    # ダイアログがVS Codeの後ろに隠れないように最前面に設定
    root.attributes('-topmost', True) 

    # フォルダ選択ダイアログを表示
    dir_path = filedialog.askdirectory(title="対象のおおもとのフォルダを選択してください")
    
    # ウィンドウを破棄（Jupyterでフリーズするのを防ぐため）
    root.destroy()
    
    return dir_path

# --- 1. フォルダの選択 ---
root_dir_path = select_directory()

# ユーザーがキャンセルした場合の処理
if not root_dir_path:
    print("フォルダの選択がキャンセルされました。処理を終了します。")
else:
    root_dir = Path(root_dir_path)
    print(f"選択されたフォルダ: {root_dir}")
    
    rename_count = 0
    print("--- 処理開始 ---")
    
    # --- 2. ファイル名の変更処理 ---
    # rglob("*")でフォルダ内のすべてのファイル・フォルダを再帰的に検索
    for file_path in root_dir.rglob("*"):
        
        # ファイルのみを対象とし、システムなどの隠しファイル（.で始まるもの）は除外
        if file_path.is_file() and not file_path.name.startswith("."):
            
            # 末端フォルダ (file_path.parent)
            # 末端からひとつ前のフォルダ (file_path.parent.parent)
            parent_folder = file_path.parent.parent
            
            # 階層が浅すぎる（おおもとのフォルダより外側の親を参照してしまう）場合はスキップ
            if root_dir in parent_folder.parents or root_dir == parent_folder:
                
                parent_name = parent_folder.name
                
                # 元のファイル名（拡張子なし）と拡張子を取得
                stem = file_path.stem       # 例: "report"
                suffix = file_path.suffix   # 例: ".xlsx"
                
                # 新しいファイル名を作成 (元のファイル名_末端からひとつ前のフォルダ名)
                new_file_name = f"{stem}_{parent_name}{suffix}"
                new_file_path = file_path.with_name(new_file_name)
                
                # すでに名前が同じになっていないかチェック
                if file_path.name != new_file_name:
                    # 変更先に同名ファイルがなければリネームを実行
                    if not new_file_path.exists():
                        file_path.rename(new_file_path)
                        print(f"変更しました: {file_path.name}  ->  {new_file_name}")
                        rename_count += 1
                    else:
                        print(f"スキップ: {new_file_name} はすでに存在します。")
            
    print("--- 処理完了 ---")
    print(f"合計 {rename_count} 件のファイル名を変更しました。")
```
